# La clase RigidBody2D: transformaciones y sólidos rígidos

La librería `rkd` cuenta con una clase `RigidBody2D` cuyo objetivo es representar gráficamente un sólido rígido definido a partir de ciertos puntos. Sobre el sólido rígido definido se pueden realizar operaciones de transformación.

Comenzamos importando la librería:

```python
from rkd.didactic import *
```

Definimos un objeto de la clase `RigidBody2D` a partir de puntos que establecen un polígono convexo.

```python
points = ((0,0),(0,1),(2,0))
r = RigidBody2D(points)
```

Para representar el sólido podemos utilizar el método `draw`:

```python
r.draw()
```

Este método puede aceptar como argumento un color, por ejemplo:

```python
r.draw("#ff00ff")
```

```python
r.draw("y")
```

Podemos ejecutar operaciones de traslación mediante el método `move`:

```python
r.move([5,10,0])
```

Para visualizar la traslación realizada podemos graficar nuevamente el sólido:

```python
r.draw("m")
```

Sería más ilustrativo si pudieramos representar ambas figuras (la original y la transformada) en una misma gráfica:

```python
r.restart() # Inicializa el sólido a su posición inicial
r.draw("r") # Dibuja el sólido en dicha posición
r.move([5,10,0]) # Movemos
r.draw("g") # volvemos graficar
```

Para realizar operaciones de rotación utilizamos el método `rotate`, el cual espera como argumento un ángulo en radianes:

```python
r.restart()
r.draw("b")
r.rotate(pi/4)
r.move([10,0,0])
r.draw("y")
```

La matriz de transformación homogénea que define la posición y orientación actual del sólido se puede obtener con el atributo `H`. El conjunto de transformaciones realizadas se puede obtener con el atributo `Hs`:

```python
r.H # Matriz de transformación homogénea del sólido
```

```python
r.Hs # Lista de transformaciones realizadas
```
